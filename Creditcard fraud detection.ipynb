```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_excel("creditcard.xlsx")

plt.figure(figsize=(12, 8))


plt.subplot(2, 2, 1)
plt.plot(df['V3'][:500], df['V4'][:500], color='red')
plt.title("Line Plot (V3 vs V4)")
plt.xlabel("V3")
plt.ylabel("V4")


plt.subplot(2, 2, 2)
pie_values = df['V4'][:5].abs()
plt.pie(pie_values, labels=df['V3'][:5], autopct="%1.1f%%")
plt.title("Pie Chart (First 5 Values)")


plt.subplot(2, 2, 3)
plt.bar(df['V3'][:10], df['V4'][:10], color='red')
plt.title("Bar Chart (10 Samples)")
plt.xlabel("V3")
plt.ylabel("V4")


plt.subplot(2, 2, 4)
plt.scatter(df['V3'][:500], df['V4'][:500], color='red')
plt.title("Scatter Plot (V3 vs V4)")
plt.xlabel("V3")
plt.ylabel("V4")

plt.tight_layout()
plt.savefig("dashboard.png",dpi=300,bbox_inches='tight')
plt.show()
```
